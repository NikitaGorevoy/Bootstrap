## Зависимость от размера выборки

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_size = pd.read_csv('CLT_norm_sample_size.csv')['sample_size'].iloc[:14]

p_hat_CLT = pd.read_csv('CLT_norm_sample_size.csv')['p_hat'].iloc[:14]
CI_upper_CLT = pd.read_csv('CLT_norm_sample_size.csv')['CI_upper'].iloc[:14]
CI_lower_CLT = pd.read_csv('CLT_norm_sample_size.csv')['CI_lower'].iloc[:14]

p_hat_Boot_percent = pd.read_csv('Percentile_norm_sample_size.csv')['p_hat'].iloc[:14]
CI_upper_Boot_percent = pd.read_csv('Percentile_norm_sample_size.csv')['CI_upper'].iloc[:14]
CI_lower_Boot_percent = pd.read_csv('Percentile_norm_sample_size.csv')['CI_lower'].iloc[:14]

p_hat_Boot_student = pd.read_csv('Studentized_norm_sample_size.csv')['p_hat'].iloc[:14]
CI_upper_Boot_student = pd.read_csv('Studentized_norm_sample_size.csv')['CI_upper'].iloc[:14]
CI_lower_Boot_student = pd.read_csv('Studentized_norm_sample_size.csv')['CI_lower'].iloc[:14]

plt.figure(figsize=(15, 8))
plt.axhline(y = 0.95, color = 'r', linestyle = '--', label='0.95')

plt.plot(sample_size, p_hat_CLT, color = 'b', label='Prob of Cover CLT')
# plt.plot(sample_size, CI_upper_CLT, color = 'b', linestyle = ':', label='CI for Prob of Cover CLT')
# plt.plot(sample_size, CI_lower_CLT, color = 'b', linestyle = ':')

plt.plot(sample_size, p_hat_Boot_percent, color = 'g', label='Prob of Cover Boot percent')
# plt.plot(sample_size, CI_upper_Boot_percent, color = 'g', linestyle = ':', label='CI for Prob of Cover Boot percent')
# plt.plot(sample_size, CI_lower_Boot_percent, color = 'g', linestyle = ':')

plt.plot(sample_size, p_hat_Boot_student, color = 'orange', label='Prob of Cover Boot student')
# plt.plot(sample_size, CI_upper_Boot_student, color = 'orange', linestyle = ':', label='CI for Prob of Cover Boot student')
# plt.plot(sample_size, CI_lower_Boot_student, color = 'orange', linestyle = ':')

# plt.title('Зависимость оцениваемой вероятности накрытия от размера выборки (нормальное распределение) \n Число экспериментов для ЦПТ = 5*10^5, для бутстрэпа = 10^5')
plt.xlabel("Sample size")
plt.ylabel("Probability")
plt.legend();
```
